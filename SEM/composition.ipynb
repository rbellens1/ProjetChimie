## 2. Fil conducteur

```python
plt.figure(figsize=(5, 6))
plt.bar(WireStr, WireWtPercent, color='skyblue')

# Ajouter des annotations pour les pourcentages
for i, percentage in enumerate(WireWtPercent):
    if percentage < 1 and percentage!= 0 :
        plt.text(i, percentage + 1, '<1%', ha='center', fontsize=10)
    else:
        plt.text(i, percentage + 1, f'{int(percentage)}%', ha='center', fontsize=10)

# Ajouter des labels et un titre
plt.xlabel('Elements')
plt.ylabel('Weight percentage (%)')
plt.title('Wire composition (SEM-EDX results)')
plt.ylim(0, max(WireWtPercent) + 10)  # Ajuster les limites de l'axe y

# Afficher le graphique
plt.tight_layout()
plt.show()

plt.figure(figsize=(5, 6))
plt.bar(Kanthal, KanthalWtPercent, color='skyblue')

# Ajouter des annotations pour les pourcentages
for i, percentage in enumerate(KanthalWtPercent):
    if percentage < 1 and percentage!= 0 :
        plt.text(i, percentage + 1, '<1%', ha='center', fontsize=10)
    else:
        plt.text(i, percentage + 1, f'{int(percentage)}%', ha='center', fontsize=10)

# Ajouter des labels et un titre
plt.xlabel('Elements')
plt.ylabel('Weight percentage (%)')
plt.title('FeCrAl composition')
plt.ylim(0, max(KanthalWtPercent) + 10)  # Ajuster les limites de l'axe y

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python

# Data for austenitic steel
austeniticSteel = ['Fe', 'Cr', 'Ni', 'Si', 'Mn', 'others']
austeniticSteelWtPercent = [[66, 74], [18, 20], [8, 11], [0, 1], [0, 2], [0, 1]]

# Calculate the average and range for each component
averages = [np.mean(percent_range) for percent_range in austeniticSteelWtPercent]
lower_bounds = [percent_range[0] for percent_range in austeniticSteelWtPercent]
upper_bounds = [percent_range[1] for percent_range in austeniticSteelWtPercent]
errors = [(avg - lower, upper - avg) for avg, lower, upper in zip(averages, lower_bounds, upper_bounds)]
asymmetric_errors = np.array(errors).T  # Transpose to separate lower and upper errors

# Create the bar plot
plt.figure(figsize=(5, 6))
bars = plt.bar(austeniticSteel, averages, color='skyblue', label='Average Weight %')

# Add whiskers (error bars) for the ranges
plt.errorbar(
    austeniticSteel, averages,
    yerr=asymmetric_errors,
    fmt='none',  # No marker for the data points
    ecolor='black',  # Color of the whiskers
    elinewidth=1.5,  # Thickness of the whiskers
    capsize=5,  # Add caps to the whiskers
    label='Range'
)

# Add annotations for the ranges
for i, (lower, upper) in enumerate(zip(lower_bounds, upper_bounds)):
    plt.text(i, upper + 1, f'{lower}-{upper}%', ha='center', fontsize=10)

# Add labels and title
plt.xlabel('Elements')
plt.ylabel('Weight Percentage (%)')
plt.title('Stainless Steel, Austenitic, AISI 304 Composition')
plt.ylim(0, max(upper_bounds) + 10)  # Adjust y-axis limits
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
